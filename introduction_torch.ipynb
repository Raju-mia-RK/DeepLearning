set random seed

```python
RANDOM_SEED=42
torch.manual_seed(RANDOM_SEED)
torch_random_C=torch.rand(3,4)
torch.manual_seed(RANDOM_SEED)
torch_random_D=torch.rand(3,4)
print(torch_random_C)
print(torch_random_D)
print(torch_random_C==torch_random_D)
```
