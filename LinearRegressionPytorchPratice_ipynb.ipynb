```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from torch import nn
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
x_numpy,y_numpy=datasets.make_regression(n_samples=100,n_features=1,noise=20,random_state=42)
```

```python
x=torch.from_numpy(x_numpy).type(torch.float)
y=torch.from_numpy(y_numpy).type(torch.float)
y=y.view(y.shape[0],1)
```

```python
n_samples,n_features=x.shape
input_size=n_features
output_size=1
```

```python
model=nn.Linear(input_size,output_size)
```

```python
criterion=nn.MSELoss()
learning_rate=0.01
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)

epochs=100

for epoch in range(epochs):
  y_pred=model(x)
  loss=criterion(y_pred,y)

  optimizer.step()
  loss.backward()
  optimizer.zero_grad()
  if(epoch+1)%10==0:
    print(f"epochs: {epoch+1} loss={loss.item():.4f}")

    predict=model(x).detach().numpy()
    plt.plot(x_numpy,y_numpy,'ro')
    plt.plot(x_numpy,predict)
    plt.show()
```
